```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as skit
import seaborn as sns
from IPython.display import display, Markdown, Latex

RAW_DATA = pd.read_csv('paaStats.csv', index_col=0).reset_index(drop=True)
df = RAW_DATA.groupby(['n'], as_index=False).mean().groupby('n')['timeBrut', 'timeHeu', 'error'].mean()
df1 = RAW_DATA.groupby(['n'], as_index=False).max().groupby('n')['error'].max()

df = pd.DataFrame(df)
display(df.head())
df1 = pd.DataFrame(df1)
display(df1.head())

df['timeBrut'] = df['timeBrut']/1000/60
df['timeHeu'] = df['timeHeu']

fig, ax = plt.subplots(figsize=(30,15))
ax.plot(df.index, df['timeBrut'] , linewidth=6)
ax.ticklabel_format(useOffset=False, style='plain')
ax.set_xticks(df.index)
ax.legend(['Brute-force'], fontsize = 50)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
#plt.title("Brute-force een. asov nronost.", size = 50)
plt.xlabel("Velikost instance v potu poloek (n)", size = 35)
plt.ylabel("as CPU na jednu instanci v minutch (min)", size = 35)
plt.grid()
plt.show()
fig.savefig("1.pdf", bbox_inches='tight')


fig, ax = plt.subplots(figsize=(30,15))
ax.plot(df.index, df['timeHeu'] , linewidth=6)
ax.ticklabel_format(useOffset=False, style='plain')
ax.set_xticks(df.index)
ax.legend(['Heuristika'], fontsize = 50)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
#plt.title("een pomoc heuristiky. asov nronost.", size = 50)
plt.xlabel("Velikost instance v potu poloek (n)", size = 35)
plt.ylabel("as CPU na jednu instanci v milisekundch (ms)", size = 35)
plt.grid()
plt.show()
fig.savefig("2.pdf", bbox_inches='tight')


fig, ax = plt.subplots(figsize=(30,30))
ax.plot(df.index, df['error'] , df1['error'], linewidth=6)
ax.ticklabel_format(useOffset=False, style='plain')
ax.set_xticks(df.index)
ax.legend(['Prmrn chyba', 'Mximln chyba'], fontsize = 50)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
#plt.title("een pomoc heuristiky. Velikost relativn chyby.", size = 50)
plt.xlabel("Velikost instance v potu poloek (n)", size = 35)
plt.ylabel("Relativn chyba v een pomoc heuristiky", size = 35)
plt.grid()
plt.show()

fig.savefig("3.pdf", bbox_inches='tight')
```
